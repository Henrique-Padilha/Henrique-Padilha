```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel("/content/varejo.xlsx")
vendas.head()
```

```python
vendas.info()
```

```python
vendas["idcanalvenda"] = vendas["idcanalvenda"].str.replace("APP","Aplicativo")
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(" ","_")
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas[vendas["estado"].isnull()]
```

```python
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```python
vendas["estado"].fillna("MS", inplace=True)
vendas.info()
```

```python
media_preco = vendas["Preço"].mean()
vendas["Preço"].fillna(media_preco, inplace=True)
```

```python
vendas.info()
```

```python
vendas[["Preço","Preço_com_frete"]].describe()
```

```python
preco_errado = vendas.query("Preço > Preço_com_frete")
vendas_correto = vendas.query("Preço < Preço_com_frete")
```

```python
vendas_correto.describe()
```

```python
vendas_correto.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
round(vendas_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(),2)
```

```python
vendas_correto.groupby("Data").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
vendas_correto["mes"] = vendas_correto["Data"].dt.month
vendas_correto.head()
```

```python
vendas_correto.groupby("mes").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
cliente = pd.read_excel("/content/cliente_varejo.xlsx")
cliente.head()
```

```python
cliente[["idade","renda"]].describe()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({"renda":"float"})
cliente.info()
```

```python
vendas_cliente = vendas_correto.merge(cliente, how= "left", on= "cliente_Log")
vendas_cliente.head()
```

```python
agg_idcanal_renda = round(vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

```python
agg_idade_bandeira = round(vendas_cliente.groupby("bandeira")["idade"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_idade_bandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(5,3))
plt.bar(agg_idade_bandeira["bandeira"], agg_idade_bandeira["idade"], color="blue")
plt.ylabel("Média de idade")
plt.title("Idade média por bandeira", loc = "left", fontsize = 14, fontweight = "bold", color = "black")
```

```python
fig = plt.figure(figsize=(5,3))
plt.bar(agg_idcanal_renda["idcanalvenda"], agg_idcanal_renda["renda"], color="blue")
plt.ylabel("Média de renda", fontweight = "bold")
plt.title("Renda média por canal de venda", loc = "left", fontsize = 14, fontweight = "bold", color = "black")
```

```python
vendas_por_data = vendas_correto.groupby("Data").idcompra.nunique().reset_index()
vendas_por_data

fig = plt.figure(figsize=(12,8))

plt.plot(vendas_por_data["Data"], vendas_por_data["idcompra"], color="blue")
plt.ylabel("Qtd de Vendas", fontweight = "bold")
plt.title("Qtd de vendas por data", loc = "center", fontsize = 14, fontweight = "bold", color = "black")
```

```python
fig = plt.figure(figsize=(12,8))

plt.xlabel("Data da venda")
plt.ylabel("Quantidade de vendas")
x = vendas_por_data["Data"]
y = vendas_por_data["idcompra"]

plt.plot(x,y, color = "blue", linewidth = "2")
plt.show()
```

```python
agg_preco_dpt = round(vendas_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_preco_dpt

fig = plt.figure(figsize=(30,10))

plt.bar(agg_preco_dpt["Nome_Departamento"], agg_preco_dpt["Preço_com_frete"], color="blue")
plt.ylabel("Preço médio", fontweight = "bold")
plt.title("Preço médio por departamento", loc = "center", fontsize = 14, fontweight = "bold", color = "black")
```
